```python
length = df["Fish_Length(cm)"]
```

```python
length
```

```python
uni = length.unique()
```

```python
uni
```

```python
import numpy as np
```

```python
length_arr = np.array(length)
```

```python
length_arr
```

```python
for y in length_arr:
    print(y)
```

```python
k_m = []

for x in uni :
    count = 0
    for y in length_arr:
        
        if(x==y):
            count = count + 1
            
        else:
            pass
    k_m.append(count)
            
    
            
            
```

```python
k_m
```

```python
len(k_m)
```

```python
len(length_arr)
```

```python
len(uni)
```

```python
mean_freq = int(np.mean(k_m))
```

```python
mean_freq
```

# The mean freq of each unique element is 2226.118

```python
key_map = {}
for A, B in zip(uni, k_m):
    key_map[A] = B
```

```python
key_map
```

```python
count_l = []
for y in length_arr:
    count_l.append(key_map[y])
```

```python
count_l=np.array(count_l)
```

```python
df["Count"] = count_l
```

```python
df.head()
```

```python
type_freq = []

for y in count_l:
    if y>mean_freq:
        type_freq.append("Slow Growth")
    else:
        type_freq.append("Fast Growth")
```

```python
type_freq
```

```python
ty_fr = np.array(type_freq)
```

```python
df["Growth type"] = ty_fr
```

```python
df.head()
```

```python
df.to_csv("df_with_freq.csv", index=False)
```
