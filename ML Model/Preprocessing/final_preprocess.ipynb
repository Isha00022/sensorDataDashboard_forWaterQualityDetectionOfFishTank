```python
import pandas as pd
import numpy as np
```

```python
df_freq = pd.read_csv("df_with_freq.csv")
df_len = pd.read_csv("df_with_length.csv")
```

```python
df_freq.head()
```

```python
df_freq.drop(["created_at", "entry_id", "Fish_Length(cm)", "Population", "Fish_Weight(g)"], axis = 1, inplace = True)
```

```python
df_freq.head()
```

```python
df_len.drop(["created_at", "entry_id", "Fish_Length(cm)", "Population", "Fish_Weight(g)"], axis = 1, inplace = True)
```

```python
df_len.head()
```

```python
df_len.drop(["Unnamed: 0"], axis = 1, inplace= True)
```

```python
df_len.head()
```

```python
df3 = df_len
```

```python
df3["Growth speed"] = df_freq["Growth type"]
```

```python
df3.head()
```

```python
final_growth = []

type_growth = df3["Type_Growth"]
growth_speed = df3["Growth speed"]
```

```python
l = len(type_growth)
```

```python
count1 = 0
count2 = 0
count3 = 0
count4 = 0
```

```python
for i in range(l):
    if type_growth[i]== "Less Growth" and growth_speed[i] == "Slow Growth" :
        final_growth.append("Less and Slow Growth")
        count1 += 1
    elif type_growth[i]== "Less Growth" and growth_speed[i] == "Fast Growth" :
        final_growth.append("Less but Fast Growth")
        count2 += 1
    elif type_growth[i]== "Good Growth" and growth_speed[i] == "Slow Growth" :
        final_growth.append("Good but Slow Growth")
        count3 += 1
    else:
        final_growth.append("Good and Fast Growth")
        count4 += 1
```

```python
count3
```

```python
df3["Final Growth"] = final_growth
```
