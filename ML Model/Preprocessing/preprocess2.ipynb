```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("file1.csv")
df.head()
```

```python
x = df.Growth_rate.unique()
```

```python
mean_x = x.mean()
```

```python
mean_x
```

```python
col = df.Growth_rate
```

```python
type_growth = []

for y in col:
    if y<mean_x:
        type_growth.append("Less Growth")
    else:
        type_growth.append("Good Growth")
```

```python
len(type_growth)
```

```python
df["Type_Growth"] = type_growth
```

```python
df.head()
```

```python
df.drop(["Growth"], axis = 1, inplace= True)
```
